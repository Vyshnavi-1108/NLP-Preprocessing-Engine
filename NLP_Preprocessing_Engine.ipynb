## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
"Love" and "love" are treated as different tokens in NLP. Without normalization, models may treat them as separate words, causing inconsistency.

### 2. What happens if stopwords are not removed?
The model may focus on less meaningful words like "is", "the", and "and", increasing noise and reducing performance.

### 3. When removing stopwords is harmful
- Sentiment analysis: "not good" → removing "not" changes meaning
- Question answering: "What is your name?" → removing "what" loses intent

### 4. Stemming vs Lemmatization
- Stemming: Cuts words (running → runn)
- Lemmatization: Converts to meaningful base form (running → run)

```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+|\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove emojis & special characters
    text = re.sub(r"[^\w\s]", " ", text)

    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Remove short words except 'no', 'not'
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)

    results.append(tokens)
```

```python
for i, tokens in enumerate(results):
    if tokens:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = sum(len(w) for w in tokens) / total

        print(f"\nSentence {i+1}")
        print("Total Tokens:", total)
        print("Unique Tokens:", unique)
        print("Average Length:", round(avg_len, 2))
```

## Analysis

- Most noisy sentence: "Win $$$ now!!! Limited offer!!!"
- Most meaningful tokens: "I am not happy with this"

```python
from collections import Counter

all_tokens = [word for sentence in results for word in sentence]

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Top 5 Least Frequent Words:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)
print(output)
```
